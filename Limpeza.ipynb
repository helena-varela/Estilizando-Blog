## Dados Tabulares

Antes de tudo, é necessário entender como os Dados são organizados na maioria dos casos.

Isso é importante para uma melhor compreensão dos exemplos que você verá a seguir.

Dados tabulares são aqueles que são organizados em tabelas, ou seja, em linhas e colunas. As linhas são registros, enquanto as colunas são os chamados "atributos"




image-188.png


Essa organização facilita a visualização e análise de tais dados.

Um conceito importante quando se trata de dados tabulares é o de "chave primária" ou "primary key".


Uma **chave primária** é um campo ou conjunto de campos numa tabela de banco de dados que identifica de forma única cada registro ou linha da tabela.

Já uma **chave estrangeira** é um campo (ou conjunto de campos) numa tabela que faz referência à chave primária de outra tabela.

No exemplo acima, CPF  e CRM_Mèdico são chaves primárias das tabelas de "paciente" e "médico",  respectivamente, enquanto CRM_Médico é uma chave estrangeira na tabela "paciente" que faz ligação com a tabela "médico".

## Uma visão mais profunda da limpeza de dados


A limpeza de dados, ou data cleansing, é um processo crucial para garantir a qualidade e a confiabilidade dos dados que serão utilizados em análises, modelos de machine learning ou outras aplicações. Envolve identificar e corrigir erros, inconsistências, duplicatas e informações irrelevantes em conjuntos de dados brutos.

Sendo a etapa inicial da análise, seu objetivo principal é preparar, ou *pré-processar*, os dados para as etapas adiante.

### Por que os dados precisam ser limpos?



Primeiro, precisamos entender o motivo de existirem "dados sujos".

Exemplos de causas de erros:
*  Falha humana.
*  Falha no processo de coleta de dados.
*  Limitações do dispositivo de medição.
*  Má fé.
*  Valor de atributo muda com o tempo.

Para que serve a limpeza?

- Obtenção de modelos mais fiéis à distribuição dos dados.
- Redução de complexidade computacional.
- Facilitar a interpretação dos padrões extraídos.

A limpeza, de fato, o ocorre apenas na etapa de limpeza?
  - A resposta é: não! O objetivo da limpeza é, de fato, pré-processar ao máximo os dados para as próximas etapas, mas, as vezes, só é possível encontrar certas incoerências nos dados quando os analisando mais profundamente, (como na etapa da Análise Exploratória, que veremos a seguir) portanto o foco é limpar ao máximo, mas "sujeiras" encontradas em etapas posteriores não significam sempre uma falha na etapa da limpeza!

### Indicativos de dados sujos

#### "Sujeiras" mais fáceis de perceber

* Dados faltantes: Ocorrem quando valores esperados estão ausentes na base de dados. Eles podem surgir por falhas de registro, erros de sensores etc.
*   Colunas sem nome: Costuma aparecer após a importação de dados mal formatados, como planilhas mal estruturadas.
*   Dados não padronizados: Valores inconsistentes dentro de uma mesma colunas, como datas em formatos diferentes ou variações de escrita ("SiM", "sim", "SIM").

Screenshot from 2025-06-07 16-29-02.png

#### "Sujeiras" mais sutis

*   Dados incorretos: Valores que não fazem sentido dentro do contexto ou que violam regras esperadas, como uma idade negativa ou uma data futura para um evento passado.
*   Colunas inúteis: Atributos que não contribuem para a análise ou modelagem, como identificadores aleatórios, colunas com valores iguais em todas as linhas ou com muitos valores únicos irrelevantes.

É de suma importância que mantenhemos em mente o contexto do nosso dataset e qual recorte estamos tentando fazer, de forma a manter todas as informações uteis à mão e eliminar todas aquelas que só irão encher o nosso dataset - e, consequentemente, aumentar a complexidade computacional em etapas posteriores

## O conceito de *Tidy Dataset*


### Organizar também é limpar

   Uma situação muito comum na etapa de limpeza é a padronização de escrita ou de tipo de variáveis. Por exemplo, pode-se ter a variável salário com os seguintes valores na tabela: 2000; 2001,00; 7500.0 etc.
  Observe que cada valor está com uma formatação diferente, o que pode dificultar a leitura na hora do código. Aqui, seria interessante converter os valores para reais e padronizar a formatação.

  Com isso, podemos ver que a organização dos dados também faz parte do pré-processamento, porém, até para a organização dos dados também precisamos seguir um padrão, caso contrário cada um organizaria os dados do seu jeito, e a organização de um pode ser a desorganização de outro!

### O que é um DataSet *Tidy*

  *Tidy* é uma forma de organização desenvolvida por Hadley Wickham, um dos principais desenvolvedores da linguagem R, para facilitar a análise estatística e a visualização, e possui as seguintes regras:

  - **Uma coluna é reservada a somente uma variável**
    - Exemplo: Uma coluna `peso / idade` deve ser dividida em duas colunas `peso` e `idade` em um dataset *Tidy*
  - **Uma linha é reservada a somente uma entrada**
    - Exemplo: Em um dataset sobre clientes de uma empresa, cada linha deve representar unicamente um cliente, corrigindo quaisquer ambiguidades.
  - **Uma tabela é reservada a somente um tipo de entrada**
    - Exemplo: Se você tem diferentes tipos de dados (ex: pacientes e exames), eles devem estar em tabelas separadas.

### Benefícios da organização *Tidy*

- Facilita a análise de dados: operações como filtragem, agrupamento e modelagem ficam mais simples.
- Integra-se melhor com bibliotecas de análise como `pandas`.
- Facilita a visualização de dados, tornando gráficos mais fáceis de gerar.
- Auxilia na hora de alimentar algum modelo de Aprendizado de Máquina com os dados organizados.

#### Funções mais básicas:

```python
# Importando as bibliotecas necessárias para a manipulação dos nossos dados
import pandas as pd
import numpy as np

# O comando gdown serve apenas para baixar arquivos do Google Drive a partir do seu ID.
!gdown 1IjWS5KSZSFzQS7jgOd8JztXgFRYGJxI9

# A função read_csv recebe o caminho de um arquivo CSV e gera um DataFrame com aqueles dados.
df = pd.read_csv('dataset_exemplos_sujo.csv')
```

#### Visualizando e manipulando dados:

print() e display()

```python
# Funções nativas do Python para visualização: print() e display()
print("Visualizando a base com o print:\n")
print(df)
print("\nVisualizando a base com o display:\n")
display(df)
```

Também podemos mostrar apenas algumas colunas ou apenas algumas linhas:

```python
# Mostrando apenas uma coluna:
display(df["Curso"])

# Mostrando apenas duas linhas (usando o operador slice):
display(df[0:2])
```

info(), describe() e shape()

```python
# info():
  # Mostra informações gerais do dataset como quantidade de colunas, quantidade de linhas, número de valores faltantes, tipo das colunas, etc.
print("Saída da função info():\n")
print(df.info())

# describe():
  # Mostra informações estatísticas do dataset, como média geral das colunas, desvio padrão geral das colunas, etc
print("\nSaída da função describe():\n")
display(df.describe())

# shape():
  # Mostra o formato do dataset: (linhas,colunas)
print("\nSaída da função shape():\n")
print(df.shape)
```

head() e tail()

```python
# função head()
  # Mostra as primeiras n linhas de um DataFrame (por padrão, 5)
print("Primeiras 9 linhas:\n")
display(df.head(9))

# função tail()
  # Mostra as últimas n linhas de um DataFrame (por padrão, 5)
print("\nÚltimas 7 linhas:\n")
display(df.tail(7))
```

loc() e iloc()

```python
# Função loc():
  # Acessa um grupo de linhas e colunas com base no nome e em um intervalo fechado:
  # Sintaxe: df.loc[(linhas), (colunas)]
print("Visualizando as colunas 'Ano_Ingresso' e 'Ano_Saida_Esperado' das primeiras 4 entradas\n")
display(df.loc[0:3, ["Ano_Ingresso", "Ano_Saida_Esperado"]])

# Função iloc():
  # Acessa um grupo de linhas e colunas com base no índice e em um intevalo aberto:
  # Sintaxe: df.iloc[(linhas), (colunas)]
print("\nMesma coisa mas usando a função iloc:")
display(df.iloc[0:4, 2:4])
```

Manipulando dados utilizando comparações básicas

```python
# Exemplo 1: Filtrar apenas as entradas de alunos que possuam bolsa
display(df[df['Possui_Bolsa'] == True].head())
```

```python
# Exemplo 2: Filtrar todos as entradas que possuam 'CRA' acima de 8
display(df[df["CRA"] > 8].head())
```

Removendo colunas

```python
# Criando uma coluna nova
df['coluna nova'] = df['ID_Aluno'] + 1
display(df.head())
```

```python
df.drop(columns=['coluna nova'], inplace=True)
display(df.head())
```

Existem também outras formas de remover colunas, como a função drop_duplicates() para duplicatas e dropna() para valores vazios.

Note que a nossa base possui alguns valores vazios, além de duplicatas:

```python
print(df.info())
```

```python
display(df.duplicated().value_counts())
print("Os valores \"True\" são duplicatas!")
```

É sempre importante ponderar sobre as entradas que estão sendo removidas, pois são ativamente informações da base de dados que estão sendo perdidas. No nosso caso, digamos que nos fosse informado préviamente que as 25 entradas que possuiam valores não fossem tão relevantes e, por isso, podemos removê-las.

```python
df.dropna(inplace=True)
df.drop_duplicates(inplace=True)
print(df.info())
```

#### Mais algumas funções interessantes:

```python
# Função transpose():
  # Transpõe os dados (inverte linhas e colunas)
print("Antes de transpor:")
display(df.head(3))
df_transposto = df.transpose()
print("\nDepois de transpor:")
display(df_transposto.head(3))
```

```python
# Funções matemáticas

df.columns = df.columns.str.strip()

# soma:
print("soma: %d" % df['CRA'].sum())

# contagem:
print("contagem: %d" % df['CRA'].count())

# média
print("média: %f" % df['CRA'].mean())

# mediana
print("mediana: %f" % df['CRA'].median())

# min
print("mínimo: %d" % df['CRA'].min())

# max
print("máximo: %d" % df['CRA'].max())

# quartil
print(df['CRA'].quantile([0.25,0.5,0.75]))

# variância
print("variância: %f" % df['CRA'].var())

# desvio padrão
print("desvio padrão: %f" % df['CRA'].std())
```

### Definindo funções num contexto de Ciência de Dados

```python
def achar_maioria(coluna):
  valores_coluna = df[coluna].value_counts()
  maioria = valores_coluna.idxmax()
  print("A maioria da coluna " + str(coluna) + " é " + str(maioria) + " com " +  str(valores_coluna.max()) + " entradas.")


achar_maioria('Possui_Bolsa')
```
